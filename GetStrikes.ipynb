```python
from swinging.ibapp import PositionsApp

def get_symbol_opt_data(symbols):
    positions_app = PositionsApp()
    positions_app.connect('127.0.0.1',  7497, clientId=0)
    positions_app.get_multiple_contract_details(symbols)
    positions_app.run()
    return positions_app.options_contract_data
```
